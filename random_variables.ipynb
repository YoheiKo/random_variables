## Generating random numbers

The SciPy package generates random variables from various distributions

### Uniform distribution
Uniform distribution (and Normal distribution) are useful in Monte Carlo simulations.<br>
The PDF for the continuous uniform distribution is: <br><br>
$ \displaystyle
             \nonumber f(x) = \left\{
              \begin{array}{l l}
                \frac{1}{b-a} &  \quad \textrm{for} \:\: a \leq x \leq b,\\\\
                0 & \quad \textrm{for} \:\: x < a \: \textrm{or} \: x > b
              \end{array} \right.
$<br>

```python
import numpy as np
from scipy.stats import uniform
uniform_array = uniform.rvs(size = 50) # generate 50 samples from uniform distribution from loc(a)=0 to scale(b-a)=1
uniform_array
```

```python
import matplotlib.pyplot as plt
plt.plot(uniform_array,marker="o", linestyle="None")
```

<li>We can see that 50 samples are uniformally distributed between 0 and 1

```python
plt.xlim(xmin=-0.5, xmax = 1.5) # set the range of x-axis
plt.hist(uniform.rvs(size = 5000)) # create histgram 
```

<li>This is a histgram of 5,000 samples

### Normal distribution
<br>
The PDF of Normal distribution is:<br><br>
$
\displaystyle
    f(x \:|\:\mu , \sigma^2) \:= \frac{1}{\sqrt{2 \pi \sigma^2}}e^{- \frac{(x-\mu)^{2}}{2 \sigma^2 }}
$
<br>

```python
from scipy.stats import norm
norm_array = norm.rvs(size = 50) # generate 50 samples from normal distribution with loc(mean)=0 and scale(std)=1
norm_array
```

```python
plt.plot(norm_array,marker="o", linestyle="None")
```

<li>50 samples are normally distributed at mean 0 and standard deviation 1

```python
plt.hist(norm.rvs(size = 5000),bins=50)
```

<li>This is a histgram of 5,000 samples. It is bell-shaped around the mean.

### Exponential distribtion

$
             \nonumber f(x; \lambda) = \left\{
              \begin{array}{l l}
                \lambda e^{-\lambda x} & \quad  x \geq 0\\
                0 & \quad \textrm{otherwise}
              \end{array} \right.
$<br>

```python
from scipy.stats import expon
expon_array = expon.rvs(size = 50,scale=1) # generate 50 samples from exponential distribution. Scale = 1 / lambda
expon_array
```

```python
plt.plot(expon_array,marker="o", linestyle="None")
```

<li>50 samples are exponentially distributed with lambda = 1

```python
plt.hist(expon.rvs(size = 5000),bins=50)
```

<li>This is a histgram of 5,000 samples.

### Multivariate normal distribution
<p>If each distribution is independent from each other(uncorrelated), multivariate normal distribution is just a matrix of one-dimensional normal distributions. If there is a non-zero correlations between distributions, we have to make use of the  correlation matrix $\sum$ </p>
<p>
To create multivariate normal distribution, we find the cholesky decomposition of the correlation matrix, and multiply the resulting lower triangular L by the matrix of uncorrelated random variables. 
</p>

```python
import numpy as np
Sigma = np.array([[1,0.5],[0.5,1]]) # This is the correlation matrix
L = np.linalg.cholesky(Sigma) # Cholesky decomposition of Sigma
uncorr_norms = norm.rvs(size = (2,20)) #2x20 matrix of uncorrelated random variables
corr_norms = np.matmul(L,uncorr_norms) # (2x2) dot (2x20) = (2x20) correlated sereis
plt.plot(corr_norms.T,marker="o")
```

<li>50 x 2 samples are normally distributed with mean = 0, std = 1, and correlation = 0.5

```python
uncorr_norms = norm.rvs(size = (2,600)) #2x20 matrix of uncorrelated random variables
corr_norms = np.matmul(L,uncorr_norms)
plt.scatter(corr_norms[0,:],corr_norms[1,:])
```

### Stochastic Process - Symmetric randomwalk
<p> Stochastic processes are random variables indexed by time. Each path is a realized sample from the probability distribution  over the spaces of paths. Symmetric Randomwalk is one of discrete-time stochastic processes. It is often comparied to a fair coin toss game. We start from zero. If the first coin toss is head, we gain one. If the first coin toss is tail, we lose one. The increment at time i is either 1 or -1:<br><br>
\begin{equation}
\displaystyle
             \nonumber X_i = \left\{
                  \begin{array}{l l}
                    \ 1 &    \: \text{if the coin toss is head},\\\\
                    -1 &  \: \text{if the coin toss is tail} 
                  \end{array}  \right.
\end{equation}<br><br>

The random walk is the sum of $X_i$ up to the time i:<br>

\begin{equation}
 M_n = \sum_{j=1}^n X_i \:\: \text{for n = 1,2,....}
\end{equation}<br><br>

Random walk starts from zero, $M_0 = 0$, its mean value is zero $\mathbb{E}(M) = 0$ and the standard deviation at time t is $\sqrt{t}$.

```python
import numpy as np
from scipy.stats import norm
import numpy.matlib
from scipy.stats import uniform
import matplotlib.pyplot as plt
import math
import random
```

### Generating randomwalk (1)
<br>
<li> 15 samples of 100 steps randomwalk generated by Monte Carlo simulation

```python
N=15 #number of simulations. 
step = 100 # steps of randomwalk

for j in range(0,N,1): # plot N numbers of  different simulations
    randomwalk_dt = np.random.randint(2,size=step)*2-1 # generate increments whih is 1 or -1 at random
    randomwalk_dt = np.insert(randomwalk_dt,0,0) # starts at zero
    randomwalk = np.cumsum(randomwalk_dt) # store the sum of increments up to the index.
    plt.plot(randomwalk) 
plt.show()
```

### Generating randomwalk (2)
<br>
<li> 30 samples of 250 steps randomwalk generated by Monte Carlo simulation.
<li> The gray lines are 2 and 3 standard deviations range.    

```python
N=30 #number of simulations. 
step = 250 # steps of randomwalk
steprange = range(0,step+1)
std = [np.sqrt(i) for i in steprange] # create empty list for storing standard deviation

for j in range(0,N,1): # plot N numbers of  different simulations
    randomwalk_dt = np.random.randint(2,size=step)*2-1 # generate increments whih is 1 or -1 at random
    randomwalk_dt = np.insert(randomwalk_dt,0,0) # starts at zero
    randomwalk = np.cumsum(randomwalk_dt) # store the sum of increments up to the index.
    plt.plot(randomwalk) 
    
plt.plot(2*np.array(std), color="lightgray") # +2std range
plt.plot(-2*np.array(std), color="lightgray") # -2std range 
plt.plot(3*np.array(std), color="gray") # +3std range
plt.plot(-3*np.array(std), color="gray") # -3std range
plt.show()
```
